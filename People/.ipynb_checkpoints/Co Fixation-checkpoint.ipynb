```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the eye-tracking data for all sessions
sed_data_path_01 = '/Users/urmebose/Desktop/People/Nuro/sed_fix_01.csv'
sed_data_path_02 = '/Users/urmebose/Desktop/People/Nuro/sed_fix_02.csv'
sed_data_path_03 = '/Users/urmebose/Desktop/People/Nuro/sed_fix_03.csv'

sed_data_01 = pd.read_csv(sed_data_path_01)
sed_data_02 = pd.read_csv(sed_data_path_02)
sed_data_03 = pd.read_csv(sed_data_path_03)

# Convert datetime formats in eye-tracking data
sed_data_01['datetime'] = pd.to_datetime(sed_data_01['datetime'], format='%Y/%m/%d %H:%M:%S.%f')
sed_data_02['datetime'] = pd.to_datetime(sed_data_02['datetime'], format='%Y/%m/%d %H:%M:%S.%f')
sed_data_03['datetime'] = pd.to_datetime(sed_data_03['datetime'], format='%Y/%m/%d %H:%M:%S.%f')

# Ensure eye-tracking data is timezone-naive
sed_data_01['datetime'] = sed_data_01['datetime'].dt.tz_localize(None)
sed_data_02['datetime'] = sed_data_02['datetime'].dt.tz_localize(None)
sed_data_03['datetime'] = sed_data_03['datetime'].dt.tz_localize(None)

# Add session labels
sed_data_01['session'] = 'Session 01'
sed_data_02['session'] = 'Session 02'
sed_data_03['session'] = 'Session 03'

# Combine datasets
all_sed_data = pd.concat([sed_data_01, sed_data_02, sed_data_03])

# Plot the comparison of fixation durations across sessions
plt.figure(figsize=(14, 7))
sns.boxplot(data=all_sed_data, x='session', y='duration')
plt.title('Comparison of Fixation Durations Across Sessions')
plt.xlabel('Session')
plt.ylabel('Fixation Duration (ms)')
plt.grid(True)
plt.tight_layout()
plt.show()

# Display descriptive statistics for fixation durations across sessions
fixation_duration_stats = all_sed_data[['session', 'duration']].groupby('session').describe()
fixation_duration_stats
```
