```python
import pandas as pd

# Load baseline HR data
baseline_hr = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr.csv')

# Load HR data for all three seasons
hr_01 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_01.csv')
hr_02 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_02.csv')
hr_03 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_03.csv')

# Function to clean HR data
def clean_hr_data(hr_data):
    hr_data_clean = hr_data[hr_data['confidence'] > 0]  # Filter out low-confidence measurements
    return hr_data_clean

# Clean the HR data
baseline_hr_clean = clean_hr_data(baseline_hr)
hr_01_clean = clean_hr_data(hr_01)
hr_02_clean = clean_hr_data(hr_02)
hr_03_clean = clean_hr_data(hr_03)

# Calculate average heart rate for baseline and each season
baseline_avg_hr = baseline_hr_clean['heart_rate'].mean()
avg_hr_01 = hr_01_clean['heart_rate'].mean()
avg_hr_02 = hr_02_clean['heart_rate'].mean()
avg_hr_03 = hr_03_clean['heart_rate'].mean()

# Calculate the difference between the baseline and each season's average HR
hr_difference_01 = avg_hr_01 - baseline_avg_hr
hr_difference_02 = avg_hr_02 - baseline_avg_hr
hr_difference_03 = avg_hr_03 - baseline_avg_hr

# Define anxiety threshold
hr_anxiety_threshold = 100

# Determine if the average HR for each season indicates anxiety
anxiety_hr_baseline = baseline_avg_hr > hr_anxiety_threshold
anxiety_hr_01 = avg_hr_01 > hr_anxiety_threshold
anxiety_hr_02 = avg_hr_02 > hr_anxiety_threshold
anxiety_hr_03 = avg_hr_03 > hr_anxiety_threshold

# Calculate Pearson's correlation coefficient for heart rate data between different periods
# Extract heart rate data for each period
baseline_hr_values = baseline_hr_clean['heart_rate']
hr_01_values = hr_01_clean['heart_rate']
hr_02_values = hr_02_clean['heart_rate']
hr_03_values = hr_03_clean['heart_rate']

# Create a DataFrame with heart rate values for correlation calculation
hr_data_for_corr = pd.DataFrame({
    'Baseline': baseline_hr_values,
    'Test_01': hr_01_values,
    'Test_02': hr_02_values,
    'Test_03': hr_03_values
})

# Calculate Pearson's correlation coefficient
correlation_matrix = hr_data_for_corr.corr(method='pearson')

# Display the results
results = {
    "Baseline Average HR": baseline_avg_hr,
    "Baseline Anxiety": "Anxiety" if anxiety_hr_baseline else "Normal",
    "Season 1 Average HR": avg_hr_01,
    "Season 1 Anxiety": "Anxiety" if anxiety_hr_01 else "Normal",
    "Difference from Baseline in Season 1": hr_difference_01,
    "Season 2 Average HR": avg_hr_02,
    "Season 2 Anxiety": "Anxiety" if anxiety_hr_02 else "Normal",
    "Difference from Baseline in Season 2": hr_difference_02,
    "Season 3 Average HR": avg_hr_03,
    "Season 3 Anxiety": "Anxiety" if anxiety_hr_03 else "Normal",
    "Difference from Baseline in Season 3": hr_difference_03
}

# Convert results to DataFrame for display
results_df = pd.DataFrame([results])
print(results_df)

print("Pearson's Correlation Coefficient Matrix:")
print(correlation_matrix)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load baseline HR data
baseline_hr = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr.csv')

# Load HR data for all three seasons
hr_01 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_01.csv')
hr_02 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_02.csv')
hr_03 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_03.csv')

# Function to clean HR data
def clean_hr_data(hr_data):
    hr_data_clean = hr_data[hr_data['confidence'] > 0]  # Filter out low-confidence measurements
    return hr_data_clean

# Clean the HR data
baseline_hr_clean = clean_hr_data(baseline_hr)
hr_01_clean = clean_hr_data(hr_01)
hr_02_clean = clean_hr_data(hr_02)
hr_03_clean = clean_hr_data(hr_03)

# Calculate average heart rate for baseline and each season
baseline_avg_hr = baseline_hr_clean['heart_rate'].mean()
avg_hr_01 = hr_01_clean['heart_rate'].mean()
avg_hr_02 = hr_02_clean['heart_rate'].mean()
avg_hr_03 = hr_03_clean['heart_rate'].mean()

# Calculate the difference between the baseline and each season's average HR
hr_difference_01 = avg_hr_01 - baseline_avg_hr
hr_difference_02 = avg_hr_02 - baseline_avg_hr
hr_difference_03 = avg_hr_03 - baseline_avg_hr

# Define anxiety threshold
hr_anxiety_threshold = 100

# Determine if the average HR for each season indicates anxiety
anxiety_hr_baseline = baseline_avg_hr > hr_anxiety_threshold
anxiety_hr_01 = avg_hr_01 > hr_anxiety_threshold
anxiety_hr_02 = avg_hr_02 > hr_anxiety_threshold
anxiety_hr_03 = avg_hr_03 > hr_anxiety_threshold

# Calculate Pearson's correlation coefficient for heart rate data between different periods
# Extract heart rate data for each period
baseline_hr_values = baseline_hr_clean['heart_rate']
hr_01_values = hr_01_clean['heart_rate']
hr_02_values = hr_02_clean['heart_rate']
hr_03_values = hr_03_clean['heart_rate']

# Create a DataFrame with heart rate values for correlation calculation
hr_data_for_corr = pd.DataFrame({
    'Baseline': baseline_hr_values.reset_index(drop=True),
    'Test_01': hr_01_values.reset_index(drop=True),
    'Test_02': hr_02_values.reset_index(drop=True),
    'Test_03': hr_03_values.reset_index(drop=True)
})

# Calculate Pearson's correlation coefficient
correlation_matrix = hr_data_for_corr.corr(method='pearson')

# Display the results
results = {
    "Baseline Average HR": baseline_avg_hr,
    "Baseline Anxiety": "Anxiety" if anxiety_hr_baseline else "Normal",
    "Season 1 Average HR": avg_hr_01,
    "Season 1 Anxiety": "Anxiety" if anxiety_hr_01 else "Normal",
    "Difference from Baseline in Season 1": hr_difference_01,
    "Season 2 Average HR": avg_hr_02,
    "Season 2 Anxiety": "Anxiety" if anxiety_hr_02 else "Normal",
    "Difference from Baseline in Season 2": hr_difference_02,
    "Season 3 Average HR": avg_hr_03,
    "Season 3 Anxiety": "Anxiety" if anxiety_hr_03 else "Normal",
    "Difference from Baseline in Season 3": hr_difference_03
}

# Convert results to DataFrame for display
results_df = pd.DataFrame([results])

# Plot the average heart rate for baseline and each season
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df.melt(id_vars=['Baseline Anxiety', 'Season 1 Anxiety', 'Season 2 Anxiety', 'Season 3 Anxiety'], 
                                 value_vars=['Baseline Average HR', 'Season 1 Average HR', 'Season 2 Average HR', 'Season 3 Average HR']), 
            x='variable', y='value')
plt.title('Average Heart Rate for Baseline and Each Season')
plt.ylabel('Average Heart Rate')
plt.xlabel('Period')
plt.xticks(rotation=45)
plt.show()

# Plot the difference from baseline heart rate for each season
plt.figure(figsize=(10, 6))
sns.barplot(data=results_df.melt(id_vars=['Baseline Anxiety', 'Season 1 Anxiety', 'Season 2 Anxiety', 'Season 3 Anxiety'], 
                                 value_vars=['Difference from Baseline in Season 1', 'Difference from Baseline in Season 2', 'Difference from Baseline in Season 3']), 
            x='variable', y='value')
plt.title('Difference from Baseline Heart Rate for Each Season')
plt.ylabel('Difference in Heart Rate')
plt.xlabel('Season')
plt.xticks(rotation=45)
plt.show()

# Print the results DataFrame and correlation matrix
print(results_df)
print("\nPearson's Correlation Coefficient Matrix:")
print(correlation_matrix)
```

```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
hr_01 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_01.csv')
hr_02 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/hr_02.csv')
hr_03 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro//hr_03.csv')

psychometric_01 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03 = pd.read_csv('/Users/urmebose/Desktop/People/Nuro/Psychometric_Test_Results_03.csv')

# Convert datetime columns to datetime objects
hr_01['datetime'] = pd.to_datetime(hr_01['datetime'])
hr_02['datetime'] = pd.to_datetime(hr_02['datetime'])
hr_03['datetime'] = pd.to_datetime(hr_03['datetime'])

psychometric_01['Question Start Time'] = pd.to_datetime(psychometric_01['Question Start Time']).dt.tz_localize(None)
psychometric_02['Question Start Time'] = pd.to_datetime(psychometric_02['Question Start Time']).dt.tz_localize(None)
psychometric_03['Question Start Time'] = pd.to_datetime(psychometric_03['Question Start Time']).dt.tz_localize(None)

psychometric_01['Question Answer Time'] = pd.to_datetime(psychometric_01['Question Answer Time']).dt.tz_localize(None)
psychometric_02['Question Answer Time'] = pd.to_datetime(psychometric_02['Question Answer Time']).dt.tz_localize(None)
psychometric_03['Question Answer Time'] = pd.to_datetime(psychometric_03['Question Answer Time']).dt.tz_localize(None)

# Calculate the answer duration for each session
psychometric_01['answer_duration'] = (psychometric_01['Question Answer Time'] - psychometric_01['Question Start Time']).dt.total_seconds()
psychometric_02['answer_duration'] = (psychometric_02['Question Answer Time'] - psychometric_02['Question Start Time']).dt.total_seconds()
psychometric_03['answer_duration'] = (psychometric_03['Question Answer Time'] - psychometric_03['Question Start Time']).dt.total_seconds()

# Merge HR and psychometric data based on the closest timestamp for each session
merged_01 = pd.merge_asof(hr_01, psychometric_01[['Question Start Time', 'answer_duration', 'Answer']], left_on='datetime', right_on='Question Start Time', direction='nearest')
merged_02 = pd.merge_asof(hr_02, psychometric_02[['Question Start Time', 'answer_duration', 'Answer']], left_on='datetime', right_on='Question Start Time', direction='nearest')
merged_03 = pd.merge_asof(hr_03, psychometric_03[['Question Start Time', 'answer_duration', 'Answer']], left_on='datetime', right_on='Question Start Time', direction='nearest')

# Drop rows with NaN values in relevant columns
merged_01_clean = merged_01.dropna(subset=['heart_rate', 'answer_duration', 'Answer'])
merged_02_clean = merged_02.dropna(subset=['heart_rate', 'answer_duration', 'Answer'])
merged_03_clean = merged_03.dropna(subset=['heart_rate', 'answer_duration', 'Answer'])

# Calculate correlations for each session
correlation_01_duration = pearsonr(merged_01_clean['heart_rate'], merged_01_clean['answer_duration'])[0]
correlation_01_score = pearsonr(merged_01_clean['heart_rate'], pd.to_numeric(merged_01_clean['Answer'], errors='coerce'))[0]

correlation_02_duration = pearsonr(merged_02_clean['heart_rate'], merged_02_clean['answer_duration'])[0]
correlation_02_score = pearsonr(merged_02_clean['heart_rate'], pd.to_numeric(merged_02_clean['Answer'], errors='coerce'))[0]

correlation_03_duration = pearsonr(merged_03_clean['heart_rate'], merged_03_clean['answer_duration'])[0]
correlation_03_score = pearsonr(merged_03_clean['heart_rate'], pd.to_numeric(merged_03_clean['Answer'], errors='coerce'))[0]

# Merge the correlations into a single DataFrame
correlations_extended = pd.DataFrame({
    'Session': ['Session 01', 'Session 02', 'Session 03'],
    'Correlation Duration': [correlation_01_duration, correlation_02_duration, correlation_03_duration],
    'Correlation Score': [correlation_01_score, correlation_02_score, correlation_03_score]
})

# Display the correlation DataFrame
print("Heart Rate vs. Answer Duration and Answer Score Correlations")
print(correlations_extended)

# Prepare the data for comparison of answer duration across sessions
# Add a column to identify the session for each dataset
psychometric_01['Session'] = 'Session 01'
psychometric_02['Session'] = 'Session 02'
psychometric_03['Session'] = 'Session 03'

# Concatenate the datasets
all_sessions = pd.concat([psychometric_01, psychometric_02, psychometric_03])

# Plot the answer duration for each session
plt.figure(figsize=(12, 6))
sns.boxplot(data=all_sessions, x='Session', y='answer_duration')
plt.title('Comparison of Answer Duration Across Sessions')
plt.ylabel('Answer Duration (seconds)')
plt.xlabel('Session')
plt.show()

# Display the descriptive statistics for answer duration across sessions
answer_duration_stats = all_sessions[['Session', 'answer_duration']].groupby('Session').describe()
print("Comparison of Answer Duration Across Sessions")
print(answer_duration_stats)
```
