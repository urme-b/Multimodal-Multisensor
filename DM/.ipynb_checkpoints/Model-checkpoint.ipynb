```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import matplotlib.pyplot as plt

# Load data
hr_01_data = pd.read_csv('/Users/urmebose/Desktop/DM/hr_01.csv')
ibi_01_data = pd.read_csv('/Users/urmebose/Desktop/DM/ibi_01.csv')
sed_01_data = pd.read_csv('/Users/urmebose/Desktop/DM/sed_01.csv')
psychometric_test_01_data = pd.read_csv('/Users/urmebose/Desktop/DM/Psychometric_Test_Results_01.csv')

hr_02_data = pd.read_csv('/Users/urmebose/Desktop/DM/hr_02.csv')
ibi_02_data = pd.read_csv('/Users/urmebose/Desktop/DM/ibi_02.csv')
sed_02_data = pd.read_csv('/Users/urmebose/Desktop/DM/sed_02.csv')
psychometric_test_02_data = pd.read_csv('/Users/urmebose/Desktop/DM/Psychometric_Test_Results_02.csv')

hr_03_data = pd.read_csv('/Users/urmebose/Desktop/DM/hr_03.csv')
ibi_03_data = pd.read_csv('/Users/urmebose/Desktop/DM/ibi_03.csv')
sed_03_data = pd.read_csv('/Users/urmebose/Desktop/DM/sed_03.csv')
psychometric_test_03_data = pd.read_csv('/Users/urmebose/Desktop/DM/Psychometric_Test_Results_03.csv')

# Clean HR data (remove entries with zero confidence)
hr_01_clean = hr_01_data[hr_01_data['confidence'] > 0]
hr_02_clean = hr_02_data[hr_02_data['confidence'] > 0]
hr_03_clean = hr_03_data[hr_03_data['confidence'] > 0]

# Aggregate HR data (example: mean heart rate)
hr_aggregated_01 = hr_01_clean.groupby('datetime')['heart_rate'].mean().reset_index()
hr_aggregated_02 = hr_02_clean.groupby('datetime')['heart_rate'].mean().reset_index()
hr_aggregated_03 = hr_03_clean.groupby('datetime')['heart_rate'].mean().reset_index()

# Clean IBI data (remove entries with zero ibi)
ibi_01_clean = ibi_01_data[ibi_01_data['ibi'] > 0]
ibi_02_clean = ibi_02_data[ibi_02_data['ibi'] > 0]
ibi_03_clean = ibi_03_data[ibi_03_data['ibi'] > 0]

# Aggregate IBI data (example: mean IBI)
ibi_aggregated_01 = ibi_01_clean.groupby('datetime')['ibi'].mean().reset_index()
ibi_aggregated_02 = ibi_02_clean.groupby('datetime')['ibi'].mean().reset_index()
ibi_aggregated_03 = ibi_03_clean.groupby('datetime')['ibi'].mean().reset_index()

# Merge HR and IBI data on 'datetime'
hr_ibi_combined_01 = pd.merge(hr_aggregated_01, ibi_aggregated_01, on='datetime')
hr_ibi_combined_02 = pd.merge(hr_aggregated_02, ibi_aggregated_02, on='datetime')
hr_ibi_combined_03 = pd.merge(hr_aggregated_03, ibi_aggregated_03, on='datetime')

# Merge HR+IBI data with ET data on 'datetime'
full_data_01 = pd.merge(hr_ibi_combined_01, sed_01_data, on='datetime')
full_data_02 = pd.merge(hr_ibi_combined_02, sed_02_data, on='datetime')
full_data_03 = pd.merge(hr_ibi_combined_03, sed_03_data, on='datetime')

# Merge all full datasets
all_data_combined = pd.concat([full_data_01, full_data_02, full_data_03], ignore_index=True)

# Print columns to verify correct names
print("Columns available in DataFrame:", all_data_combined.columns)

# Define features based on actual column names you confirmed
features = [
    'heart_rate', 'ibi', 'headPos_x', 'headPos_y', 'headPos_z', 
    'gazeDir_x', 'gazeDir_y', 'gazeDir_z', 'pupil_diameter',  
    'HADS_A', 'HADS_D', 'STAI_S', 'STAI_T', 'BFI', 'FQ'        
]

# Filter out features that are missing in the DataFrame
existing_features = [feature for feature in features if feature in all_data_combined.columns]

# Print missing columns
missing_features = set(features) - set(existing_features)
if missing_features:
    print(f"Missing columns in the DataFrame: {missing_features}")

# Ensure features exist in the data, drop rows with missing values in these columns
all_data_combined = all_data_combined.dropna(subset=existing_features)

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(all_data_combined[existing_features])

# Perform PCA to reduce dimensionality for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Determine the optimal number of clusters using the silhouette score
silhouette_scores = []
K = range(2, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    score = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(K, silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of Clusters')
plt.show()

# Choose the optimal number of clusters (e.g., k=3 based on silhouette scores)
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Visualize the clusters in PCA-reduced space
plt.figure(figsize=(10, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', marker='o')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-Means Clusters (k={}) in PCA-Reduced Space'.format(optimal_k))
plt.colorbar(label='Cluster')
plt.show()

# Display cluster centers and sizes
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_sizes = pd.Series(clusters).value_counts()

print("Cluster Centers:\n", cluster_centers)
print("\nCluster Sizes:\n", cluster_sizes)

# Calculate and display clustering evaluation metrics
silhouette_avg = silhouette_score(X_scaled, clusters)
dbi = davies_bouldin_score(X_scaled, clusters)
chi = calinski_harabasz_score(X_scaled, clusters)

print(f"Silhouette Score: {silhouette_avg}")
print(f"Davies-Bouldin Index: {dbi}")
print(f"Calinski-Harabasz Index: {chi}")

# Convert silhouette score to a percentage for interpretation
silhouette_percentage = (silhouette_avg + 1) / 2 * 100
print(f"Clustering 'accuracy' (Silhouette Score as a percentage): {silhouette_percentage:.2f}%")
```
